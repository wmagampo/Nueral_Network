```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_file = "./dog.jpg"

img = cv2.imread(image_file, 0) / 255
```

```python
plt.imshow(img)
```

```python
# Resize image
img_width = 28
img_height = 28

dim = (img_width, img_height)

img = cv2.resize(img, dim)

plt.imshow(img)
```
